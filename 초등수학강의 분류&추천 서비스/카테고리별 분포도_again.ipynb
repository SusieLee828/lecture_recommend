```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import matplotlib.font_manager as fm
from matplotlib import rc
import seaborn as sns
```

```python
# Windows의 경우 Malgun Gothic 사용
font_path = 'C:/Windows/Fonts/malgun.ttf'
fontprop = fm.FontProperties(fname=font_path)
# 전역 폰트 설정
rc('font', family=fontprop.get_name())
```

```python
# 'labels' 컬럼이 문자열일 때만 eval 적용, 이미 리스트면 그대로 사용
def convert_to_list(x):
    if isinstance(x, str):
        try:
            return eval(x)  # 문자열이면 eval로 리스트로 변환
        except:
            return x  # eval이 실패하면 원본 그대로 반환
    return x  # 이미 리스트나 다른 자료형이면 그대로 반환

# 데이터 불러오기
data = pd.read_csv('category1.csv')

# 'labels' 컬럼에 변환 함수 적용
data['category'] = data['category'].apply(convert_to_list)

# 'labels' 컬럼을 펼쳐서 빈도 계산
data_exploded = data.explode('category')

# 'khan'과 'ebs'로 데이터 나누기
khan_data = data_exploded[data_exploded['source'] == 'khan']
ebs_data = data_exploded[data_exploded['source'] == 'ebs']

# 각 레이블 빈도 계산
khan_label_counts = Counter(khan_data['category'])
ebs_label_counts = Counter(ebs_data['category'])

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.set(font_scale=1.3)

# 막대 위에 숫자 표시하는 함수
def add_value_labels(bars, ax):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va='bottom', ha='center', fontsize=10)

# Khan 그래프
khan_labels, khan_counts = zip(*khan_label_counts.items())
bars_khan = axes[0].bar(khan_labels, khan_counts, color='skyblue')
axes[0].set_title('Khan Academy 카테고리별 분포', fontproperties=fontprop, fontsize=18, weight='bold')
axes[0].set_xlabel('카테고리', fontproperties=fontprop, fontsize=15, weight='bold')
axes[0].set_ylabel('총 개수', fontproperties=fontprop, fontsize=15, weight='bold')
axes[0].tick_params(axis='x')
add_value_labels(bars_khan, axes[0])

# EBS 그래프
ebs_labels, ebs_counts = zip(*ebs_label_counts.items())
bars_ebs = axes[1].bar(ebs_labels, ebs_counts, color='lightcoral')
axes[1].set_title('EBS 카테고리별 분포', fontproperties=fontprop, fontsize=18, weight='bold')
axes[1].set_xlabel('카테고리', fontproperties=fontprop, fontsize=15, weight='bold')
axes[1].tick_params(axis='x')

# 막대 위에 숫자 표시
add_value_labels(bars_ebs, axes[1])

# 레이아웃 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
# 'labels' 컬럼이 문자열일 때만 eval 적용, 이미 리스트면 그대로 사용
def convert_to_list(x):
    if isinstance(x, str):
        try:
            return eval(x)  # 문자열이면 eval로 리스트로 변환
        except:
            return x  # eval이 실패하면 원본 그대로 반환
    return x  # 이미 리스트나 다른 자료형이면 그대로 반환

# 데이터 불러오기
data = pd.read_csv('category1.csv')

# 'labels' 컬럼에 변환 함수 적용
data['category'] = data['category'].apply(convert_to_list)

# 'labels' 컬럼을 펼쳐서 빈도 계산
data_exploded = data.explode('category')

# 'khan'과 'ebs'로 데이터 나누기
khan_data = data_exploded[data_exploded['source'] == 'khan']
ebs_data = data_exploded[data_exploded['source'] == 'ebs']

# 각 레이블 빈도 계산
khan_label_counts = Counter(khan_data['category'])
ebs_label_counts = Counter(ebs_data['category'])

all_categories = sorted(set(khan_label_counts.keys()).union(set(ebs_label_counts.keys())))

khan_counts_ordered = [khan_label_counts.get(cat, 0) for cat in all_categories]
ebs_counts_ordered = [ebs_label_counts.get(cat, 0) for cat in all_categories]

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.set(font_scale=1.3)

# 막대 위에 숫자 표시하는 함수
def add_value_labels(bars, ax):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va='bottom', ha='center', fontsize=10)

# Khan 그래프
bars_khan = axes[0].bar(all_categories, khan_counts_ordered, color='lightcoral')
axes[0].set_title('Khan Academy 카테고리별 분포', fontproperties=fontprop, fontsize=18, weight='bold')
axes[0].set_xlabel('카테고리', fontproperties=fontprop, fontsize=15, weight='bold')
axes[0].set_ylabel('총 개수', fontproperties=fontprop, fontsize=15, weight='bold')
axes[0].tick_params(axis='x')
add_value_labels(bars_khan, axes[0])

# EBS 그래프
bars_ebs = axes[1].bar(all_categories, ebs_counts_ordered, color='skyblue')
axes[1].set_title('EBS 카테고리별 분포', fontproperties=fontprop, fontsize=18, weight='bold')
axes[1].set_xlabel('카테고리', fontproperties=fontprop, fontsize=15, weight='bold')
axes[1].tick_params(axis='x')

# 막대 위에 숫자 표시
add_value_labels(bars_ebs, axes[1])

# 레이아웃 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
